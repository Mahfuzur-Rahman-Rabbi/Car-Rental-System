```python
class Car:
    def __init__(self, company, model, year, available=True):
        self.company = company
        self.model = model
        self.year = year
        self.available = available

    def rent(self):
        if self.available:
            self.available = False
            print(f"{self.company} {self.model} {self.year} has been rented.")
        else:
            print(f"{self.company} {self.model} {self.year} is not available for rent.")

    def return_car(self):
        if not self.available:
            self.available = True
            print(f"{self.company} {self.model} {self.year} has been returned.")
        else:
            print(f"{self.company} {self.model} {self.year} is already available.")


class CarRentalSystem:
    def __init__(self):
        self.cars = []

    def add_car(self, car):
        self.cars.append(car)

    def display_available_cars(self):
        print("Available Cars:")
        for car in self.cars:
            if car.available:
                print(f"{car.company} {car.model} {car.year}")

    def rent_car(self, company, model, year):
        for car in self.cars:
            if car.company == company and car.model == model and car.year == year:
                car.rent()
                return
        print("Car not found.")

    def return_car(self, company, model, year):
        for car in self.cars:
            if car.company == company and car.model == model and car.year == year:
                car.return_car()
                return
        print("Car not found.")

car1 = Car("Ford", "Focus", 2018)
car2 = Car("Honda", "Civic", 2019)
car3 = Car("Toyota", "Corolla", 2020)
car4 = Car("Toyota", "Noah Hybrid", 2023)
car5 = Car("Nissan", "Xtrail", 2024)
car6 = Car("Toyota", "Hiace", 2024)

car_rental_system = CarRentalSystem()
car_rental_system.add_car(car1)
car_rental_system.add_car(car2)
car_rental_system.add_car(car3)
car_rental_system.add_car(car4)
car_rental_system.add_car(car5)
car_rental_system.add_car(car6)

while True:
    print("\n1. Display Available Cars")
    print("2. Car Rental Price")
    print("3. Rent a Car")
    print("4. Return a Car")
    print("5. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        car_rental_system.display_available_cars()
    elif choice == "2":
        print("Rental list:")
        print("Ford Focus 2018 300tk per hour")
        print("Honda Civic 2019 400tk per hour")
        print("Toyota Corolla 2020 500tk per hour")
        print("Toyota Noah Hybrid 2023 800tk per hour")
        print("Nissan Xtrail 2024 1000tk per hour")
        print("Toyota Hiace 2024 1000tk per hour")
    elif choice == "3":
        company = input("Enter car company: ")
        model = input("Enter car model: ")
        year = int(input("Enter car year: "))
        car_rental_system.rent_car(company, model, year)
    elif choice == "4":
        company = input("Enter car company: ")
        model = input("Enter car model: ")
        year = int(input("Enter car year: "))
        car_rental_system.return_car(company, model, year)
    elif choice == "5":
        print("Exiting...")
        break
    else:
        print("Invalid choice. Please try again.")
```
